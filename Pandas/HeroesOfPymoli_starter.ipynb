* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
df = pd.DataFrame(purchase_data)
df.head()
```

```python
df = df.rename(columns={'Purchase ID': 'Purchase_ID',
                       'Item ID': 'Item_ID',
                       'Item Name': 'Item_Name'})
```

```python
#******************
u_users = df['SN'].nunique()
u_users
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#********
u_items = df['Item_ID'].nunique()
avg_price = df['Price'].mean()
num_purch = df['Purchase_ID'].nunique()
tot_rev = df['Price'].sum()
```

```python
#********
summary_df = pd.DataFrame({'Number of Unique Items':[u_items], 
                           'Average Price':avg_price, 
                           'Number of Purchases':num_purch,
                           'Total Revenue':tot_rev})
summary_df['Average Price'] = summary_df['Average Price'].map("${:,.2f}".format)
summary_df['Total Revenue'] = summary_df['Total Revenue'].map("${:,.2f}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#************
gender_df = df.groupby(['Gender']).SN.nunique().sort_values(ascending = False).reset_index().rename(columns={'SN': 'Total Count'})
```

```python
#*************
gender_df['Percent of Players']=(gender_df['Total Count']/u_users)
gender_df['Percent of Players'] = gender_df['Percent of Players'].map('{:,.2%}'.format)
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purch_df = df.groupby(['Gender']).agg({'Purchase_ID':'nunique',
                                             'Price':['mean',sum],
                                              'SN':'nunique'
                                             }).reset_index()
gender_purch_df.columns = ["_".join(x) for x in gender_purch_df.columns.ravel()]
```

```python
gender_purch_df['new'] = gender_purch_df['Price_sum'] / gender_purch_df['SN_nunique']
gender_purch_df
```

```python
gender_purch_df=gender_purch_df.drop(columns = ['SN_nunique'])
```

```python
g_df = gender_purch_df.rename(columns = {'Gender_':'Gender',
                                 'Purchase_ID_nunique':'Purchase Count',
                                 'Price_mean':'Average Purchase Price',
                                 'Price_sum':'Total Purchase Value',
                                  'new':'Avg Total Purchase per Person'
                                 })
```

```python
g_df['Average Purchase Price'] = g_df['Average Purchase Price'].map("${:,.2f}".format)
g_df['Total Purchase Value'] = g_df['Total Purchase Value'].map("${:,.2f}".format)
g_df['Avg Total Purchase per Person'] = g_df['Avg Total Purchase per Person'].map("${:,.2f}".format)
g_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,1000]
bins
```

```python
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
```

```python
df["Age Bin"] = pd.cut(df["Age"], bins, labels=labels)
```

```python
age_df = df.groupby(['Age Bin']).SN.nunique().reset_index().rename(columns={'SN': 'Total Count'})
age_df
```

```python
age_df['Percent of Players']=(age_df['Total Count']/u_users)
age_df['Percent of Players'] = age_df['Percent of Players'].map('{:,.2%}'.format)
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purch_df = df.groupby(['Age Bin']).agg({'Purchase_ID':'nunique',
                                             'Price':['mean',sum],
                                              'SN':'nunique'
                                             }).reset_index()
age_purch_df.columns = ["_".join(x) for x in age_purch_df.columns.ravel()]
```

```python
age_purch_df['new'] = age_purch_df['Price_sum'] / age_purch_df['SN_nunique']
age_purch_df
```

```python
age_purch_df = age_purch_df.drop(columns = ['SN_nunique'])
```

```python
a_df = age_purch_df.rename(columns = {'Age Bin_':'Age Bin',
                                 'Purchase_ID_nunique':'Purchase Count',
                                 'Price_mean':'Average Purchase Price',
                                 'Price_sum':'Total Purchase Value',
                                  'new':'Avg Total Purchase per Person'
                                 })
```

```python
a_df['Average Purchase Price'] = a_df['Average Purchase Price'].map("${:,.2f}".format)
a_df['Total Purchase Value'] = a_df['Total Purchase Value'].map("${:,.2f}".format)
a_df['Avg Total Purchase per Person'] = a_df['Avg Total Purchase per Person'].map("${:,.2f}".format)
a_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
ts_df = df.groupby(['SN']).agg({'Purchase_ID':'nunique',
                                             'Price':['mean',sum]
                                             }).reset_index()
ts_df.columns = ["_".join(x) for x in ts_df.columns.ravel()]
```

```python
ts_df = ts_df.sort_values(ascending = False, by=['Price_sum']).head().rename(columns = {'SN_':'SN',
                                 'Purchase_ID_nunique':'Purchase Count',
                                 'Price_mean':'Average Purchase Price',
                                 'Price_sum':'Total Purchase Value'
                                 }).set_index('SN')
ts_df['Average Purchase Price'] = ts_df['Average Purchase Price'].map("${:,.2f}".format)
ts_df['Total Purchase Value'] = ts_df['Total Purchase Value'].map("${:,.2f}".format)
ts_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
mp_df = df.groupby(['Item_ID', 'Item_Name']).agg({'Purchase_ID':'nunique',
                                             'Price':['mean',sum]
                                             }).reset_index()
mp_df.columns = ["_".join(x) for x in mp_df.columns.ravel()]
```

```python
mp_df = mp_df.sort_values(ascending = False, by=['Purchase_ID_nunique']).head().rename(columns = {'Item_ID_':'Item ID',
                                                                                          'Item_Name_':'Item Name',
                                 'Purchase_ID_nunique':'Purchase Count',
                                 'Price_mean':'Item Price',
                                 'Price_sum':'Total Purchase Value'
                                 }).set_index(['Item ID','Item Name'])

mp_df['Item Price'] = mp_df['Item Price'].map("${:,.2f}".format)
mp_df['Total Purchase Value'] = mp_df['Total Purchase Value'].map("${:,.2f}".format)
mp_df
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
mpp_df = df.groupby(['Item_ID', 'Item_Name']).agg({'Purchase_ID':'nunique',
                                             'Price':['mean',sum]
                                             }).reset_index()
mpp_df.columns = ["_".join(x) for x in mpp_df.columns.ravel()]
mpp_df.head()
```
